```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing image as nparray
def load_image(filename) :
    image = Image.open(filename)
    image.load()
    img = np.asarray(image, dtype="int32")
    return img
```

```python
# Displaying an image
def show_image(array):
    return plt.imshow(array)
```

```python
# Image shape : img.shape
```

```python
# Reshaping the image
def reshape_image(array):
    X = array.reshape(224 * 224, 3)
    return X
```

```python
# Example
img = load_image('IMG-20230617-WA0012.jpeg')
show_image(img);
```
