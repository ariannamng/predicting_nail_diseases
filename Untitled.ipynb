```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing image as nparray
def load_image(filename) :
    image = Image.open(filename)
    image.load()
    img = np.asarray(image, dtype="int32")
    return img
```

```python
# Displaying an image
def show_image(array):
    return plt.imshow(array)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Opening the image with CV2
image = cv2.imread('raw_data/Onychomykose.jpg')

# Converting it to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Converting it to HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Natural nails colours
lower_bound = np.array([0, 30, 60])  # Hue, Saturation, Value pour les teintes roses ples
upper_bound = np.array([20, 150, 255])

# Creating a mask for colours within that range
mask = cv2.inRange(image_hsv, lower_bound, upper_bound)

# Selecting the right part
image_detouree = cv2.bitwise_and(image_rgb, image_rgb, mask=mask)

# Plotting the selection
plt.imshow(image_detouree)
plt.axis('on')
plt.show()
```
